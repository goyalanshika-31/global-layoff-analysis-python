```python
# Importing necessary libraries for data analysis and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick  # for formatting axis
import seaborn as sns

# Make charts look better
sns.set(style="whitegrid")
# To display plots inside notebook
%matplotlib inline
```

```python
# Reading the uploaded CSV file into a DataFrame
df = pd.read_csv('/content/Layoffs_Dataset.csv')
```

```python
# Quick look
df.head()
df.info()
```

```python
# Check column names
df.columns

# Check basic info (data types, null values)
df.info()

# Check missing values in each column
df.isnull().sum()
```

```python
# Function to clean Layoff Count column
def convert_layoff(val):

    # Handle missing or unknown values
    if pd.isna(val) or str(val).lower() == "unknown":
        return np.nan

    val = str(val)  # convert everything to string

    # Handle ranges with "–" OR "-"
    if "–" in val or "-" in val:
        val = val.replace("–", "-")   # make both same
        nums = val.split("-")
        return (int(nums[0]) + int(nums[1])) / 2

    # Handle values like "100+"
    if "+" in val:
        return int(val.replace("+",""))

    # Ignore percentage values
    if "%" in val:
        return np.nan

    # Convert normal values
    return int(val)

# Apply function
df['Layoff_Count_Num'] = df['Layoff Count'].apply(convert_layoff)
```

```python
# Convert 'Date of Layoff' into datetime format
df['Date_of_Layoff'] = pd.to_datetime(
    df['Date of Layoff'],
    errors='coerce',
    format='%b-%y'
)

# Handle cases where only year is given
df['Date_of_Layoff'] = df['Date_of_Layoff'].fillna(
    pd.to_datetime(df['Date of Layoff'], format='%Y', errors='coerce')
)
```

```python
# Check cleaned dataset
df.head()

# Check if any null values still exist
df.isnull().sum()
```

                            Exploratory Data Analysis(EDA)

```python
#Top Companies by Layoffs
# Group by company and calculate total layoffs
top_companies = df.groupby('Company Name')['Layoff_Count_Num'].sum().sort_values(ascending=False)

# Show top 10
top_companies.head(10)
```

                                                                 
                                                                 
                                                            Visualization

```python
# Plot top companies
plt.figure(figsize=(12,6))

sns.barplot(
    x=top_companies.head(10).values,
    y=top_companies.head(10).index,
    color= 'purple'
)

plt.title("Top 10 Companies by Layoffs")
plt.xlabel("Number of Employees")
plt.ylabel("Company")

plt.show()
```

```python
# Total layoffs by industry
industry_layoffs = df.groupby('Industry')['Layoff_Count_Num'].sum().sort_values(ascending=False)
top10 = industry_layoffs.head(10)

# Plot
plt.figure(figsize=(10,5))

ax= sns.barplot(
    x=top10.index,
    y=top10.values,
    color= 'blue'
)

plt.xticks(rotation=45)
plt.title("Layoffs by Industry")
plt.ylabel("Total Layoffs")

# Format Y-axis
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))


plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Layoffs by city
city_layoffs = df.groupby('City')['Layoff_Count_Num'].sum().sort_values(ascending=False)
top10 = city_layoffs.head(10)


# Plot
plt.figure(figsize=(10,5))

sns.barplot(
    x=top10.index,
    y=top10.values,
    color ='purple'
)

plt.xticks(rotation=45)
plt.title("Layoffs by City")

plt.show()
```

```python
# Layoffs based on funding status
funding = df.groupby('Funding Status')['Layoff_Count_Num'].sum()

# Plot
plt.figure(figsize=(6,4))

sns.barplot(
    x=funding.index,
    y=funding.values,
    color='blue'
)

plt.title("Layoffs by Funding Status")

plt.show()
```

```python
# Group by month
trend = df.groupby(df['Date_of_Layoff'].dt.to_period('M'))['Layoff_Count_Num'].sum()

# Convert index to proper date
trend.index = trend.index.to_timestamp()

# Plot
plt.figure(figsize=(10,5))

trend.plot(marker='o')

plt.title("Layoff Trend Over Time")
plt.xlabel("Month-Year")
plt.ylabel("Total Layoffs")

plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
# Get top 10 reasons
top10_reasons = df['Reason'].value_counts().head(10).index

# Plot
plt.figure(figsize=(12,6))

sns.countplot(
    y='Reason',
    data=df,
    order=top10_reasons,   # only top 10
    color='purple'
)

plt.title("Top 10 Reasons for Layoffs")

plt.show()
```
